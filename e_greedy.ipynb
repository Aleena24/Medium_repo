```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
# Ice cream shops and their true reward probabilities
true_probabilities = [0.74, 0.95, 0.45]  # Probabilities of getting a reward from each shop
n_shops = len(true_probabilities)  # Number of ice cream shops

# Total rounds (days) to try ice cream shops
n_rounds = 1000

# Epsilon value (chance of exploring a new shop)
epsilon = 0.1

# Track total rewards and shop selections
rewards = np.zeros(n_shops)  # Total rewards from each shop
selections = np.zeros(n_shops)  # Number of times each shop is selected

# Simulate trying shops for n_rounds (100 days)
for day in range(n_rounds):
    if random.uniform(0, 1) < epsilon:
        # Explore: randomly choose any shop
        shop_choice = random.randint(0, n_shops - 1)
    else:
        # Exploit: choose the shop with the most rewards
        shop_choice = np.argmax(rewards)

    # Simulate reward (1 for success, 0 for failure) from the shop
    reward = np.random.binomial(1, true_probabilities[shop_choice])

    # Update total rewards and selections
    rewards[shop_choice] += reward
    selections[shop_choice] += 1

# Calculate the total rewards received
total_rewards_received = rewards.sum()

# Find the best shop based on total rewards
best_shop = np.argmax(rewards) + 1
print(f"\nBest shop to visit: Shop {best_shop}")
print(f"Total rewards received: {total_rewards_received}")
```

```python
# Simple bar chart to show selections and rewards
shops = ['Creamy Delights', 'Scoops of Joy', 'Chill Treats']

# Plot shop selections (how many times each shop was chosen)
plt.figure(figsize=(12, 5))

# Plot selections
plt.subplot(1, 2, 1)
plt.bar(shops, selections, color=['skyblue', 'lightgreen', 'salmon'])
plt.title('Number of Times Each Shop Was Chosen')
plt.ylabel('Selections')

# Plot rewards (how much reward each shop gave)
plt.subplot(1, 2, 2)
plt.bar(shops, rewards, color=['skyblue', 'lightgreen', 'salmon'])
plt.title('Total Rewards From Each Shop')
plt.ylabel('Total Rewards')

# Display total rewards received on the plot
for i, v in enumerate(rewards):
    plt.text(i, v + 0.05, str(int(v)), ha='center', va='bottom')

# Show total rewards received
plt.suptitle(f'Total Rewards Received: {total_rewards_received}', fontsize=14)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
