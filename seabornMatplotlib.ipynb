```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
iris = sns.load_dataset("iris")
iris.head()
```

# Visualizations with Matplotlib

Scatter plot

```python
# Matplotlib Scatter Plot
plt.figure(figsize=(8, 5))
plt.scatter(iris['sepal_length'], iris['sepal_width'], c='blue', alpha=0.5)
plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.grid(True)
plt.show()
```

Histogram

```python
# Matplotlib Histogram
plt.figure(figsize=(8, 5))
plt.hist(iris['petal_length'], bins=20, color='green', alpha=0.7)
plt.title("Distribution of Petal Length")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.show()
```


Strengths of Matplotlib:

Total control over the plots.

Suitable for highly customized plots.


# Visualizations with Seaborn

Pair Plot

A pair plot displays pairwise relationships between features.

```python
# Seaborn Pair Plot
sns.pairplot(iris, hue="species", height=2.5)
plt.show()
```

Box Plot

Box plots help in visualizing the distribution and identifying outliers.

```python
# Seaborn Box Plot
plt.figure(figsize=(8, 5))
sns.boxplot(x="species", y="petal_length", data=iris, palette="Set2")
plt.title("Petal Length Distribution by Species")
plt.show()
```

Heatmap

A heatmap visualizes the correlation between features.

```python
# Seaborn Heatmap
plt.figure(figsize=(8, 5))

# Select only numeric columns for correlation
numeric_iris = iris.select_dtypes(include=['float64', 'int64'])
correlation = numeric_iris.corr()

# Plot the heatmap
sns.heatmap(correlation, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

Strengths of Seaborn:

Effortlessly creates complex and beautiful visualizations.

Focused on statistical plots.

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species", style="species")
plt.title("Combined Plot: Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.grid(True)
plt.show()
```
